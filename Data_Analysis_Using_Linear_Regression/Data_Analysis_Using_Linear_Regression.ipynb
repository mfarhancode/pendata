Understanding Linear Regression

Linear regression is a data analysis technique that predicts unknown data values using related, known data values. It predicts future events (continued values). Examples: population size, stock levels.

### Implementation using program

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Setup Data
X = np.array([[2], [4], [3], [3], [3], [4], [5]])
y = np.array([2, 3, 5, 4, 3, 5, 6])

# 2. Instantiate and Fit Library Model
model = LinearRegression()
model.fit(X, y)

# Print calculations
print(f"Intercept (b0): {model.intercept_:.2f}")
print(f"Slope Coefficient (b1): {model.coef_[0]:.2f}")
print(f"Equation: y = {model.intercept_:.2f} + {model.coef_[0]:.2f}x")

# 3. IMAGE / PLOT FEATURE

# Plot the blue data points
plt.scatter(X, y, color='blue', label='Data Points', s=80)

# Generate predictions for a smooth line spanning from x=1 to x=6
x_range = np.linspace(1, 6, 100).reshape(-1, 1)
y_pred = model.predict(x_range)

# Plot the red regression line
plt.plot(x_range, y_pred, color='red', label=f'y = {model.intercept_:.2f} + {model.coef_[0]:.2f}x')

# Chart labels and settings
plt.title("Linear Regression Visualization")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

# Display the image
plt.show()
```
